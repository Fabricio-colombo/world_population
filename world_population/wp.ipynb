# Análise de dados sobre a população mundial

```python
#inicio do projeto, analise de dados sobre a população mundial, com base em dados do banco mundial.
#trouxemos o arquivo > world_population_data.csv < para o projeto, e vamos analisar os dados contidos nele.

#primeiro de tudo é importar as bibliotecas que vamos usar ao longo da analise.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#agora vamos importar o arquivo csv para o projeto, e vamos analisar os dados contidos nele.

df = pd.read_csv('world_population_data.csv')
display(df)
```

```python
#analisando o grafico, percebemos que temos algumas colunas indesejadas para nossa analise, 
#sendo elas (cca3, area (km²), density (km²), growth rate ), vamos excluir elas

df.drop(columns=['cca3', 'area (km²)', 'density (km²)', 'growth rate'], inplace=True)
display(df)
```

```python
df.info()
```

```python
#vamos transformar >  12  world percentage  234 non-null    object  < em >  12  world percentage  234 non-null    float64  <

df['world percentage'] = df['world percentage'].str.replace(',', '.').str.replace('%', '').astype(float)
df.info()
```

# Qual é o continente mais populoso do mundo atualmente?

```python
#para resposta essa pergunta. Qual é o continente mais populoso do mundo atualmente?, vamos agrupar a coluna continent e somar a coluna population 2023.

continent_population = df.groupby('continent')['2023 population'].sum()
display(continent_population)
```

```python
# Criar uma figura com dois subplots (1 linha, 2 colunas)
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Primeiro subplot - Gráfico de barras
axes[0].bar(continent_population.index, continent_population.values, color=['red', 'blue', 'green', 'yellow', 'orange', 'purple'])
axes[0].set_title('População por continente (Barras)')
axes[0].set_xlabel('Continente')
axes[0].set_ylabel('População')

# Segundo subplot - Gráfico de pizza
axes[1].pie(continent_population, autopct='%1.1f%%', colors=['red', 'blue', 'green', 'yellow', 'orange', 'purple'])
axes[1].set_title('População por continente (Pizza)')
axes[1].set_xlabel('Continente')
axes[1].set_ylabel('População')

# Ajuste de layout para evitar sobreposição
plt.tight_layout()

# Exibir os subplots
plt.show()
```

# Quais os paises mais populosos do mundo atualmente?

```python
# Selecionar os 20 países mais populosos e formatar a população com separador de milhão
top_20_countries = df[['country', '2023 population']].sort_values(by='2023 population', ascending=False).head(20)
top_20_countries['2023 population'] = top_20_countries['2023 population'].apply('{:,.0f}'.format)

# Exibir os resultados
display(top_20_countries)
```

```python
# Criar um gráfico de linhas com a linha azul transparente
plt.figure(figsize=(16, 8))
plt.plot(top_20_countries['country'], top_20_countries['2023 population'], marker='o', linestyle='-', color='b', alpha=0.5)
plt.title('População dos 20 Países Mais Populosos (2023)')
plt.xlabel('País')
plt.ylabel('População')
plt.xticks(rotation=45, ha='right')  # Rotacionar os rótulos do eixo x para melhor legibilidade

# Adicionar rótulos de valor aos pontos
for i, txt in enumerate(top_20_countries['2023 population']):
    plt.annotate(txt, (top_20_countries['country'].iloc[i], top_20_countries['2023 population'].iloc[i]), ha='center')

# Adicionar linhas conectando os marcadores ao eixo x
for i, (country, population) in enumerate(zip(top_20_countries['country'], top_20_countries['2023 population'])):
    plt.vlines(country, population, 0, linestyles='dashed', color='gray', alpha=0.5)
    plt.plot([country, country], [population, 0], linestyle='dashed', color='gray', alpha=0.5)

# Inverter a ordem do eixo y
plt.gca().invert_yaxis()

plt.tight_layout()
plt.show()
```

# Qual foi o crescimento populacional de 1970 até os dias atuais?

```python
#agora queremos saber sobre (2023 population, 2022 population, 2020 population, 2015 population, 2010 population, 2000 population, 1990 population, 1980 population, 1970 population)
#vamos agrupar essas colunas e somar elas, para saber a população mundial em cada ano. separando o numero por casa de milhar.

world_population = df[['2023 population', '2022 population', '2020 population', '2015 population', '2010 population', '2000 population', '1990 population', '1980 population', '1970 population']].sum()
world_population = world_population.apply('{:,.0f}'.format)
display(world_population)
```

```python
import numpy as np

# Criar uma escala de cores do verde forte para o verde claro
colors = plt.cm.Greens(np.linspace(0.2, 1, len(world_population)))

# Criar o gráfico de barras horizontais com a escala de cores
plt.figure(figsize=(16, 8))
bars = plt.barh(world_population.index[::-1], world_population.values[::-1], color=colors)
plt.title('População Mundial (1970 - 2023)')
plt.xlabel('População')
plt.ylabel('Ano')

# Adicionar rótulos de valor nas barras
for bar, value in zip(bars, world_population.values[::-1]):
    plt.text(bar.get_width() - 0.00 * bar.get_width(), bar.get_y() + bar.get_height() / 2, f'{value}', va='center', color='black')

plt.show()
```

# Análise do Crescimento Populacional Mundial

 De 1970 até 2023, a população mundial saiu de 3.7 bilhões para 8 bilhões, um aumento de 4.3 bilhões de pessoas.

Então, a média anual de crescimento da população mundial é aproximadamente 81.132 milhões de pessoas por ano, e a média por década é aproximadamente 811.321 milhões de pessoas por década.

Pode não parecer evidente à primeira vista no gráfico, mas uma análise mais profunda da população mundial revela um fenômeno interessante: embora o número absoluto de habitantes cresça, a taxa percentual de crescimento por pessoa diminui.

Por exemplo, no período de 1970 a 1980, observamos um aumento populacional de 3.6 bilhões para 4.4 bilhões, o que representa um acréscimo de 800 milhões. Isso implica em um crescimento médio de aproximadamente 80 milhões por ano, considerando o ponto de partida de 3.6 bilhões.

Ao analisarmos os dados mais recentes, como de 2022 para 2023, percebemos que a média continua similar, saindo de 7.9 bilhões para 8 bilhões em apenas um ano. Entretanto, o número atual de habitantes é quase o dobro do que era na década de 1970. Isso nos leva a concluir que, proporcionalmente, o crescimento deveria ser o dobro do observado naquela época.

Essa constatação nos faz questionar se, em alguns anos, o gráfico tenderá a se equilibrar e, possivelmente, apresentar uma redução na população mundial.

A pergunta que permanece é: existe um limite máximo sustentável de habitantes no planeta? Se a resposta for afirmativa, qual seria esse limite? É uma indagação complexa que envolve não apenas aspectos demográficos, mas também considerações ambientais, econômicas e sociais.

A busca por respostas para essa questão é crucial para o planejamento futuro e a sustentabilidade do nosso planeta. O equilíbrio entre o crescimento populacional e os recursos disponíveis torna-se um desafio global que exige reflexão e ações cuidadosas.
